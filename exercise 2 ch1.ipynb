```python
import networkx as nx
```

```python
G=nx.Graph()
nodes_to_add=['b','c','d']
#G.add_nodes_from(nodes_to_add)
G.add_edge('a','b')
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges_to_add)
nx.draw(G,with_labels=True,node_color='yellow',node_size=1600,font_color='red',font_size=16)
```

```python
l=[]
def max_degree(g):
    x=0
    y=''
    for node in g.nodes:
        if x < g.degree(node):
            x=g.degree(node)
            y=node
    l.append(('name',y))
    l.append(('degree',x))
    return l            
print(max_degree(G))
```
